```python
details = {
    'jayanth': {
        'age': 20,
        'gender': 'male',
        'subjects': ['maths', 'python', 'ml']
    },
    'vijay': {
        'age': 22,
        'gender': 'male',
        'subjects': ['physics', 'chemistry', 'biology']
    }}
```

```python
new=[]
for i in details.keys():
    for j in details[i].values():
     print(j)
     if type(j)==list:
         for k in j:
            if k=='python':
                new.append(i)
                
```
